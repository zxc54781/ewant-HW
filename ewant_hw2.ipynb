# HW2_3

```python
def inverse (x):
    str="" #由於要用來連接必須空白不然會跑出奇怪的東西
    A=[ ]   
    x=list(x)
    n=len(x) #算字串
    for i in range (1,(n+1)):
        a=x[-i]
        A.append(a) #把結果存到Ａ
    print(str.join(A)) #印出來
```

```python
inverse('賽甲國中')
```

# HW2_4

```python
def 天干地支 (x):
    str=""
    A=[ ]
    S=list('甲乙丙丁戊己庚辛壬癸')
    G=list('子醜寅卯辰巳午未申酉戌亥')
    s=(x-2014)%10
    g=(x-2014+6)%12
    A.append(S[s])
    A.append(G[g])
    print(' {}年 是 {}年！'.format(x,str.join(A)))

```

```python
天干地支(2014)
```

```python
天干地支(2073)
```

```python
for i in range (2014,2074):
    天干地支(i)
```
