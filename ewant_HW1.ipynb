## HW1-3

```python
Equipment1 = 100
Equipment2 = 150
Equipment3 = 230
Equipment4 = 80

M1 = Equipment1
M2 = M1 + Equipment2
M3 = M2 + Equipment3
M4 = M3 + Equipment4
M=[1,2,3,4]

spend =[M1,M2,M3,M4]
```

```python

plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(M,spend,label='累積花費')
plt.xlabel('時間(月)')
plt.ylabel('花費(k)')
plt.title('累積花費')
plt.xticks(np.arange( 1,5))

plt.figure(figsize=(10,10))    # 顯示圖框架大小
plt.subplot(2,1,2)
Equiment =  'Equipment1','Equipment2','Equipment3','Equipment4'     # 製作圓餅圖的類別標籤
separeted = (0, 0, 0.3, 0, 0.3)                  # 依據類別數量，分別設定要突出的區塊
size = [100,150,230,80]                        # 製作圓餅圖的數值來源

plt.pie(size,                          # 數值
        labels =Equiment, # 標籤      
        autopct='%1.1f%%' )                    # 設定陰影

 
plt.axis('equal')                                          # 使圓餅圖比例相等
plt.title("平均花費", {"fontsize" : 18})  # 設定標題及其文字大小

plt.show()
```

* 用簡單的random解決選擇性障礙

```python
import random

a = input ('是否要輸入選項名稱(yes or no)：')
item = input ('有幾個選項：')
n=int(item)
A=[]

if a =="no":
    items=random.randint(1,n)
    print("建議你選第"+str(items)+'個選項的餐廳' )
elif a=="yes":
    for i in range(1,n+1):
        #B.insert(i,A)
        i = input('第'+str(i)+'項：')
        A.append(i)
    ans= random.choice(A)
    print("我覺得今天吃『"+str(ans)+'』是你最好的選擇')
    
        
else :
    print('你到底想怎樣！')
    
```
